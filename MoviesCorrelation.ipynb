```python
# Import Libraries 

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']= (12,8) # Adjusts the configuration of the plots we will create

# Read in the data

df=pd.read_csv('https://raw.githubusercontent.com/deepkh3/datasets/main/movies.csv')

# Looking at the data

df.head()




```

```python
# Missing data?

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%' .format(col, pct_missing))
    
```

```python
df=df.dropna()
```

```python
df.dtypes
```

```python
df['gross'] = df['gross'].astype('int64')
df['budget'] = df['budget'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
df.head()
```

```python
#Create correct year
df['yearcorrect'] = df['released'].astype(str).str.split(',').str[-1].astype(str).str[:5]
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
df
```

```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# correlation
# scatter plot with budget vs gross
plt.figure(figsize=(15,15))
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('budget')
plt.ylabel('gross')
plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

plt.figure(figsize=(15,15))
sns.regplot(x='budget', y = 'gross', data = df, scatter_kws = {"color" : "red"}, line_kws={"color" : "blue"})
```

```python
# Looking at correlation

df.corr(method = 'pearson')
```

```python
# Heat map
plt.figure(figsize=(15,15))
correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot=True, cmap= sns.diverging_palette(220,20,n=200))
plt.show()
```

```python
#Look at Company

df_numerized = df
for col_name in df_numerized.columns:
  if(df_numerized[col_name].dtype == 'object'):
    df_numerized[col_name] = df_numerized[col_name].astype('category')
    df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized

```

```python
plt.figure(figsize=(15,15))
correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot=True, cmap= sns.diverging_palette(220,20,n=200))
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
